```python
import pandas as pd
```

```python
df=pd.read_excel('FR007利率互换收盘曲线（全）(1).xls',index_col=0)[1:]
df.index.name='Time'
df
```

```python
new_col_name=[]
for col in df.columns:
#     print(col)
    col=col.replace('利率互换收盘曲线_均值','')
    col=col.replace('中债国债到期收益率','国债')
    col=col.replace('中债国开债到期收益率','国开')
    col=col.replace('中债进出口行债到期收益率','口行')
    
    col=col.replace('个月','M')
    col=col.replace('年','Y')
    new_col_name.append(col)
df.columns=new_col_name
df.index=list(map(lambda x:str(x)[:10],df.index.tolist()))
df.index.name='Time'
df
```

```python
import re 
main_time=[x for x in df.columns if re.search('1Y|2Y|3Y|:5Y|7Y|10Y',x)]
new=df.loc[:,main_time].dropna()
# new.to_csv('各基准利率数据.csv')
new.to_csv('../data/IRS/各基准利率数据.csv')
```
