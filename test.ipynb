```python
import akshare as ak
import pandas as pd
```

```python

bond_china_yield_df = ak.bond_china_yield(start_date="2021-1-05", end_date="2021-1-05")
bond_china_yield_df.shape
```

```python
day='2021-01-05'
# import akshare as
import datetime
today=datetime.datetime.strptime(day,'%Y-%m-%d')
today
```

```python
today_line=bond_china_yield_df[bond_china_yield_df['曲线名称']=='中债中短期票据收益率曲线(AAA)'].iloc[0]
today_line=today_line.iloc[2:-1]
today_line.index=[3/12,6/12,1,3,5,7,10]
today_line
```

```python
day_line=pd.Series(index=[0.25,0.50]+[i for i in range(1,11)],dtype=float)
for one in day_line.index:
    if one in today_line.index:
        day_line.loc[one]=today_line[one]
    else:
        year0=(today_line[today_line.index<one].index[-1])
        year1=(today_line[today_line.index>one].index[0])
        rate0=(today_line[today_line.index<one].iloc[-1])
        rate1=(today_line[today_line.index>one].iloc[0])
        day_line.loc[one]=round((rate0+(one-year0)/(year1-year0)*(rate1-rate0)),4)
day_line
```

```python
deal_df=pd.read_excel('./data/BondDeal/中票.xlsx')
deal_df
```

```python
small_df=deal_df
import re
def pred_rate(year):
    if year<1 and year<0.5:
        y0=0.25
        y1=0.5
    if year<1 and year>0.5:
        y0=0.5
        y1=1
    else:
        y0=int(year)
        y1=y0+1
#     print(y0)
#     print(year)
    rate=day_line.loc[y0]+(year-y0)*(day_line.loc[y1]-day_line.loc[y0])
    return rate
small_df=small_df[(small_df['day']/365>0.5)&(small_df['day']/365<5)]
small_df['real_ytm']=small_df['收益率'].apply(lambda x:float(re.findall('\d+\.\d+|(?<!-)\d+',x)[0]))
small_df['比较基准']=(small_df['day']/365).apply(lambda x:pred_rate(x))
small_df['利差']=small_df['real_ytm']-small_df['比较基准']

# small_df=small_df[(small_df['利差']<3)] # 利差小于300bp的
# small_df[0]=0

# sns.scatterplot(small_df['剩余期限'],small_df['利差'])
small_df
```
